```python
def read_nxt_lines(fh, num):
    count = 0
    result = []
    while count < num:
        res_l = fh.readline()
        if len(res_l) > 0:
            result.append(res_l)
        else:
            break
        count += 1
    return result
```

```python
def last_n_lines(file_name, num=10):
    fh = open(file_name)
    # we will read 10 lines using read_nxt_lines(fh, 10)
    prev = []
    nxt = []
    while True:
        nxt = read_nxt_lines(fh, 10)
        if len(nxt) < num:
            break
        else:
            prev = nxt
    fh.close()
    l = len(nxt)
    diff = num - l
    return prev[-diff:] + nxt
```

```python
# Q1: Find the longest line from a file?
def longest_line(filename):
    fh = open(filename)
    max_len, longest_line = 0, ''
    for line in fh:
        if len(line) > max_len:
            longest_line = line
            max_len = len(line)
    fh.close()
    return longest_line
```

```python
# Q1: Find the top 3 longest lines from a file?
def longest_k_line(filename):
    fh = open(filename)
    top3lines = [] #0th will shortest, 2nd will be longest
    for line in fh:
        if len(top3lines) < 3:
            top3lines.append((len(line), line))
            top3lines.sort()
        if len(line) > top3lines[0][0]:
            top3lines[0] = (len(line), line)
            top3lines.sort()
    fh.close()
    return top3lines
```

Q: Write a function that takes a filename and prints all the lines starting with a prefix.

```python
def get_starting_with(filename, prefix):
    with open(filename) as fh:
        for line in fh:
            if line.startswith(prefix):
                print(line.strip())
```

```python
def save_grep(input_file, keyword, out_file):
    inpf = open(input_file)
    outf = open(out_file, "a")
    keyword = keyword.lower()
    for line in inpf:
        line1 = line.lower()
        if keyword in line1:
            outf.write(line)
    outf.close()
    inpf.close()
```

```python
save_grep('mbox-short.txt', 'From:', 'mbox-results12344.txt')
```

```python
#Q: from a list of keywords, if either keyword is found in the line, save that line to another file
def save_grep_multi(input_file, keywords, out_file):
    inpf = open(input_file)
    outf = open(out_file, "w")
    for line in inpf:
        for keyword in keywords:
            if keyword in line:
                outf.write(line)
                break
    outf.close()
    inpf.close()
```

```python
save_grep_multi('mbox-short.txt', ['From', 'Subject'], 'output-mbox-multi.txt')
```
